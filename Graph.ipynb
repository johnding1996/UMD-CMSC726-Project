#### Create Random Graphs

Mainly 5 types of random graphs:
- 1st type ER model: equal prob for graphs with $n$ nodes and $m$ edges, cannot ensure connected
- 2st type ER model: $n$ nodes, $p$ prob for each edge independently, cannot ensure connected
- WS model: $n, m$ ring lattice, revise edges with prob $k$, cannot ensure connected, need node order to find prob but cannot retrieve the order
- NWS model: $n, m$ ring lattice, add edges with prob $k$, can ensuer connected, need node order to find prob, and can retrieve the order
- BA model: $n$ node and min degree $m$, growing algorithm, can ensure connected, need node order to find prob but cannot retrieve the order

Thus only consider NWS model now

```python
def create_graphs(N, n, k, p):
    return [nx.newman_watts_strogatz_graph(n, k, p) for _ in range(N)]

nx.draw(create_graphs(5, 5, 2, 0.3)[0])
```
